# Визуализация набора

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

def show(img_path, boxes, rect_th=2, text_size=1.2, text_th=2):
 
    img = cv2.imread(img_path) 
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    for i in range(len(boxes)):
        cv2.rectangle(img, (boxes[i][0], boxes[i][1]), (boxes[i][2], boxes[i][3]), color=(0, 255, 0), thickness=rect_th) 
        cv2.putText(img, 'Face', (boxes[i][0], boxes[i][1]),  cv2.FONT_HERSHEY_SIMPLEX, text_size, (255,0,0),thickness=text_th) 
    plt.figure(figsize=(10,10)) 
    plt.imshow(img)
    plt.show()
```

```python
import os
import json 

lists = os.listdir('dataset/label_Faster_R_CNN/')
```

```python
print(lists[0:10])
print(lists[3][:-5])
```

```python
for label in lists[0:10]:
    with open('dataset/label_Faster_R_CNN/' + label, "r") as file:
        metk = json.load(file)  
    
    img_path = 'dataset/img_Faster_R_CNN/'+label[:-5]+'.jpg'
    boxes = metk["box"]
    print(boxes)
    
    show(img_path, boxes)
```
