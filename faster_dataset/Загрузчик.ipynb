```python
import os 
import json

txt_fales = os.listdir('FDDB-folds')
paths = txt_fales[1::2]
labels = txt_fales[0::2]
```

```python
f = 0

for path, label in zip(paths, labels):    
    metki = []
    with open('FDDB-folds/' + label, "r") as file:
        line = file.readline()
        while line:
            metki.append(line[:-1])
            line = file.readline()
    path_img = []
    with open('FDDB-folds/' + path, "r") as file:
        line = file.readline()
        while line:
            path_img.append(line[:-1])
            line = file.readline()
    
    index = []
    for i, label_ in enumerate(metki):
        if label_ in path_img:
            index.append(i)
        
    index1 = index[:-1]
    index2 = index[1:]
    label_list = []
    for ind1, ind2 in zip(index1, index2):
        label_list.append(metki[ind1:ind2])
    
    for label in label_list:
        num_fase = int(label[1])
        katalog_img = label[0]
        box_rect = []
        for j in range(num_fase):
            tochki = label[j+2][:-3].split(' ')
            x_center = round(float(tochki[-2]))
            y_center = round(float(tochki[-1]))
            y_rad = round(float(tochki[0]))
            x_rad = round(float(tochki[1]))
            x_1 = max(x_center - x_rad, 0)
            x_2 = max(x_center + x_rad, 0)
            y_1 = max(y_center - y_rad, 0)
            y_2 = max(y_center + y_rad, 0)
            box_rect.append([x_1,y_1,x_2,y_2])
        print(box_rect)
        os.rename(katalog_img + '.jpg','img_Faster_R_CNN/' + katalog_img.split('/')[-1] + '.jpg')
        os.rename('img_Faster_R_CNN/' + katalog_img.split('/')[-1] + '.jpg',
                  'img_Faster_R_CNN/' + str(f) +'.jpg')
        end_label = {'num_fase':num_fase,
                     'box':box_rect}
        with open('label_Faster_R_CNN/'+str(f)+'.json', "w") as file:
             json.dump(end_label, file)        
        f = f + 1
```
